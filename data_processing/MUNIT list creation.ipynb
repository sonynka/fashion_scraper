```python
import pandas as pd
```

```python
df = pd.read_csv('../../../data/fashion/img_attr.csv', encoding='utf-8')
df.head()
```

```python
floral = df.loc[df.pattern_floral == 1, 'img_path'].tolist()
not_floral = df.loc[df.pattern_floral == 0, 'img_path'].sample(len(floral)).tolist()
```

```python
floral_file = open('floral.txt', 'w')
for item in floral:
    floral_file.write("%s\n" % item)
floral_file.close()
```

```python
not_floral_file = open('not_floral.txt', 'w')
for item in not_floral:
    not_floral_file.write("%s\n" % item)
not_floral_file.close()
```
