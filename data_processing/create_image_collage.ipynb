```python
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
data_dir = '/Users/sonynka/HTW/MasterArbeit/Report/pix2pix_results/28062018_dressestomodels/test_latest/images/'
files = sorted(os.listdir(data_dir))
files = [os.path.join(data_dir, f) for f in files[142:] if f.endswith('.png')]
```

```python
len(files)/3
```

```python
fig = plt.figure(figsize=(9, (len(files)/3)*3))
for idx, img_path in enumerate(files):
    ax = fig.add_subplot(len(files)/3, 3, idx+1)
    ax.axis('off')
    img = Image.open(img_path)
    plt.imshow(img)
plt.tight_layout()
plt.axis('off')
plt.show()
fig.savefig(os.path.join(data_dir, 'test_images3.jpg'))
```
